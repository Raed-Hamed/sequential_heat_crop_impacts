# Aggregate variables on county level and save in common csv file

```python
import os
import sys
import pathlib
from pathlib import Path
import json
import xarray as xr
import numpy as np

dir_preprocessing = os.path.join(str(pathlib.Path().resolve()))
sys.path.append(dir_preprocessing)

import functions_counties as fct
```

```python
geo_area = 'eu'

with open(f'config_{geo_area}.json', 'r') as file:
    config_data = json.load(file)
```

```python
crops_dict = config_data['study_area']['crops_dict']
crops = list(config_data['study_area']['crops_dict'].keys())
print(crops)
```

```python
t_annual = xr.open_dataset(config_data['calc_his']['annual_file'].format(variable='tasmax'))
sm_annual = xr.open_dataset(config_data['calc_his']['annual_file'].format(variable='SMroot'))
```

```python
ds_var_list = [t_annual, sm_annual] 
var_abbrev = ['T','M'] 

seasons = ['spring', 'summer']
years = np.arange(config_data['study_area']['his_years'][0], config_data['study_area']['his_years'][1])
```

```python
for crop in crops:
    print(crop)
    melted_list = fct.compute_counties_hist(config_data, ds_var_list, var_abbrev, crop, seasons, years)
    final_data = fct.create_historic_csv(melted_list, crop, config_data, True)
```

```python
fct.plot_counties(final_data, 1980, 'T', 'spring', 'maize', config_data['study_area']['lon'], config_data['study_area']['lat'])
```
